# Metrics

#### Imports

```python
from kilroyplot.plot import plt
import pandas as pd
```

#### Load data

```python
df = pd.read_csv("metrics.csv")
```

#### Results

```python
def plot_real_scores(df, x_label="Iteration", y_label="Score", alpha=0.5):
    df = df[df["id"] == "realScore"]
    grouped = df.groupby("x")
    x = [x for x, _ in grouped]
    y = grouped["y"].median()
    lower = grouped["y"].quantile(0.25).values
    upper = grouped["y"].quantile(0.75).values
    plt.plot(x, y, marker="o", markersize=6, linestyle="--", dashes=(2, 2))
    plt.fill_between(x, lower, upper, alpha=alpha)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
```

```python
plot_real_scores(df)
```
